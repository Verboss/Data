```python
#Dodawanie nowej kolumny (dodaly mi sie 3 przez przypadek) :)
df['Average'] = (df.Open + df.Close) / 2.0
df.head()
```

```python
#odwrócona kolejność
df = df.sort_index()
df.head()
```

```python
#przesuwamy dane "w dół" o 1
df[['Close']].shift(1)
```

```python
df['Close_shift'] = df.Close.shift(1)
df.head()
```

```python
#zmiana ceny zakniecia wyrazona w procentach (w nowej kolumnie)
df['Daily_Change'] = df.Close / df.Close_shift - 1
df.head()
```

```python
#możliwosć usuwania całych kolumn
del df['average']
del df['verage']
del df['Daily Change']
df.head()
```

```python
#najmniejsza wartość z kolumny
df.Daily_Change.min()
```

```python
#najwieksza wartość z kolumny
df.Daily_Change.max()
```
