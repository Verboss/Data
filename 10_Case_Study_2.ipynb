```python
import numpy as np
import pandas as pd
```

```python
df_raw = pd.read_csv('appstore.csv')
df_raw.head()
```

```python
df_raw.columns
```

```python
#kopiujemy dane do donej zmiennej
df = df_raw.copy()
df.info()
```

```python
df = df.drop(columns=['URL', 'Icon URL', 'Subtitle', 'Description', 'In-app Purchases'])
df.head()
```

```python
df = df.set_index('ID')
df
```

```python
pd.set_option('display.float_format', lambda x: f'{x:.2f}')
df.describe()
```

```python
df.describe().T
```

```python
#wyświetlamy wszystkie dane typu 'object' czyli tekstowe
df.describe(include=['object'])
```

```python
#transponowanie
df.describe(include=['object']).T
```

```python
df.columns
```

```python
#zmieniamy spacje w kolumnach na podkreslniki
df.columns = [col.lower().replace(' ','_') for col in df.columns]
df
```
